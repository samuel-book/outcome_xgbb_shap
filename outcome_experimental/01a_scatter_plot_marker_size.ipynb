# Scatter plot marker size

What size value represents the relative value difference as marker size.

```python
# Turn warnings off to keep notebook tidy
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

import time
```

Report the time duration to run notebook

```python
start_time = time.time()
```

## Dummy data

```python
list_y_pred = [1,1,2,2]
list_y_test = [1,2,1,2]
count_categories = np.array([1,2,4,8])
```

```python
np.sqrt(count_categories)
```

## Plot using different size representation

```python
# Scatter plot and regression line
count_categories = np.array([1,2,4,8])

fig = plt.figure(figsize=(18,6))
ax1 = fig.add_subplot(131)
ax1.scatter(list_y_pred, list_y_test, edgecolors='black', facecolor='black', alpha=0.5, 
            marker='o', s=count_categories*500)
ax1.grid()
ax1.set_xlabel('Predicted value')
ax1.set_ylabel('Observed value')
ax1.set_xlim((0.5,2.5))
ax1.set_ylim((0.5,2.5))

counter = 0
# To add values to plot
for i, j in zip(list_y_pred, list_y_test):
    plt.text(i, j, count_categories[counter], ha='center', va='center', color='w')
    counter += 1

count_categories1 = count_categories**2
ax2 = fig.add_subplot(132)
ax2.scatter(list_y_pred, list_y_test, edgecolors='black', facecolor='black', alpha=0.5, 
            marker='o', s=count_categories1*500)
ax2.grid()
ax2.set_xlabel('Predicted value')
ax2.set_ylabel('Observed value')
ax2.set_xlim((0.5,2.5))
ax2.set_ylim((0.5,2.5))

counter = 0
# To add values to plot
for i, j in zip(list_y_pred, list_y_test):
    plt.text(i, j, count_categories[counter], ha='center', va='center', color='w')
    counter += 1

count_categories2 = np.sqrt(count_categories)
ax2 = fig.add_subplot(133)
ax2.scatter(list_y_pred, list_y_test, edgecolors='black', facecolor='black', alpha=0.5, 
            marker='o', s=count_categories2*500)
ax2.grid()
ax2.set_xlabel('Predicted value')
ax2.set_ylabel('Observed value')
ax2.set_xlim((0.5,2.5))
ax2.set_ylim((0.5,2.5))

counter = 0
# To add values to plot
for i, j in zip(list_y_pred, list_y_test):
    plt.text(i, j, count_categories[counter], ha='center', va='center', color='w')
    counter += 1

plt.show()
```

```python
end_time = time.time()

print(f'Time taken: {end_time - start_time}')
```
